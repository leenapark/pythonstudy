```python
fileRead = open(fileName, encoding="UTF-8")
data = csv.reader(fileRead)
header = next(data)
# print(header)
fields = []
yangcheon = []

# title 추출
for title in header[3:]:
    fields.append(title.split("_")[-1])
print(fields)

# 연령별 데이터
for row in data:
    if "신정4동" in row[0]:
        for i in row[3:]:
            # print(row)
            yangcheon.append(int(i.replace(",", "")))
print(yangcheon)
fileRead.close()
```

```python
plt.style.use("ggplot")
# .plot(x축 데이터, y축 데이터)
plt.plot(fields, yangcheon)
plt.show()
```

```python
town = input("동네 이름을 넣어주세요 ex.{}".format(town))

def townPlt(townName):
    fileRead = open(fileName, encoding="UTF-8")
    data = csv.reader(fileRead)
    header = next(data)
    # print(header)
    fields1 = []
    yangcheon1 = []

    # title 추출
    for title in header[3:]:
        fields1.append(title.split("_")[-1])
    print(fields1)

    # 연령별 데이터
    for row in data:
        if townName in row[0]:
            for i in row[3:]:
                # print(row)
                yangcheon1.append(int(i.replace(",", "")))
    print(yangcheon1)
    fileRead.close()

    plt.style.use("ggplot")
    # .plot(x축 데이터, y축 데이터)
    plt.stem(fields1, yangcheon1)
    plt.show()
    
townPlt(town)
```

```python
seoulData = "./data_analysis/seoul_data.csv"
seoulRead = open(seoulData, encoding="UTF-8")
seoulData = csv.reader(seoulRead)
# print(seoulData)
seoulInfo = pd.DataFrame(seoulData)
# print("borough: ", headerSeoul.values[:, 0])
# print("head: ", headerSeoul.head(1))
# print("borough: ", headerSeoul.values[1, :])
# head = headerSeoul.head(1)
# peopleSeoul = seoulData
# print("peopleSeoul: ", peopleSeoul)
fieldsSeoul = []
seoulPp = []

# title 추출
for borough in seoulInfo.values[:, 0]:
    if "서울특별시" in borough and "구" in borough:
        # print(borough)
        # print(borough.split(" ")[1])
        fieldsSeoul.append(borough.split(" ")[1])
# print(fieldsSeoul)

for population in seoulInfo.values[:, 1]:
    # print(population)
    if "," in population:
        # print(int(population.replace(",", "")))
        seoulPp.append(int(population.replace(",", "")))

del seoulPp[0:1]
# print(seoulPp)

plt.style.use("ggplot")
# .plot(x축 데이터, y축 데이터)
plt.plot(fieldsSeoul, seoulPp)
plt.show()
```
