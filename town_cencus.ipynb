```python
fileRead = open(fileName, encoding="UTF-8")
data = csv.reader(fileRead)
header = next(data)
# print(header)
fields = []
yangcheon = []

# title 추출
for title in header[3:]:
    fields.append(title.split("_")[-1])
print(fields)

# 연령별 데이터
for row in data:
    if "신정4동" in row[0]:
        for i in row[3:]:
            # print(row)
            yangcheon.append(int(i.replace(",", "")))
print(yangcheon)
fileRead.close()
```

```python
plt.style.use("ggplot")
# .plot(x축 데이터, y축 데이터)
plt.plot(fields, yangcheon)
plt.show()
```

```python
plt.style.use("ggplot")
# .plot(x축 데이터, y축 데이터)
plt.step(fields, yangcheon)
plt.show()
```

```python
plt.style.use("ggplot")
# .plot(x축 데이터, y축 데이터)
plt.stem(fields, yangcheon)
plt.show()
```

```python
town = input("동네 이름을 넣어주세요 ex.{}".format(town))

def townPlt(townName):
    fileRead = open(fileName, encoding="UTF-8")
    data = csv.reader(fileRead)
    header = next(data)
    # print(header)
    fields1 = []
    yangcheon1 = []

    # title 추출
    for title in header[3:]:
        fields1.append(title.split("_")[-1])
    print(fields1)

    # 연령별 데이터
    for row in data:
        if townName in row[0]:
            for i in row[3:]:
                # print(row)
                yangcheon1.append(int(i.replace(",", "")))
    print(yangcheon1)
    fileRead.close()

    plt.style.use("ggplot")
    # .plot(x축 데이터, y축 데이터)
    plt.stem(fields1, yangcheon1)
    plt.show()
    
townPlt(town)
```
